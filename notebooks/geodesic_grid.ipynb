```python
import plotly.graph_objects as go
import plotly.io as pio
from mdgp.utils import sphere_uniform_grid


pio.templates.default = "plotly_dark"
```

```python
n = 200  # Number of points
points = sphere_uniform_grid(n)

trace = go.Scatter3d(
    x=points[:, 0],
    y=points[:, 1],
    z=points[:, 2],
    mode='markers',
    marker=dict(
        size=5,
        color=list(range(len(points))),
        opacity=0.8
    )
)

layout = go.Layout(
    title='Approximately Uniform Grid on Unit Sphere',
    scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z')
    )
)

fig = go.Figure(data=[trace], layout=layout)
fig.show()
```
